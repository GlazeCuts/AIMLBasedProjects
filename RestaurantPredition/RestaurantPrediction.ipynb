```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score, accuracy_score,precision_score,recall_score
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("D:\VScode\CODES\AIMLBasedProjects\RestaurantPredition\Dataset .csv")
```

```python
#cuisine have 9 missing values
df['Cuisines'].fillna(df['Cuisines'].mode()[0])
```

```python
#task 1: Predicting the aggregate rating of a restaurant based on other features

#Feature Selection
features_rating = ['Has Table booking', 'Has Online delivery', 'Price range', 'Votes']
target_rating = 'Aggregate rating'

#Encoding categorical variable
le = LabelEncoder()
for col in ['Has Table booking', 'Has Online delivery']:
    df[col] = le.fit_transform(df[col])

```

```python
#performing splitting
X_rating = df[features_rating]
y_rating = df[target_rating]
X_train_rating,X_test_rating,y_train_rating,y_test_rating = train_test_split(X_rating,y_rating)
```

```python
dtr = DecisionTreeRegressor(random_state=42) #random variable
dtr.fit(X_train_rating,y_train_rating)

```

```python
#Model Evaluation
y_pred_rating = dtr.predict(X_test_rating)
mse = mean_squared_error(y_test_rating, y_pred_rating)
r2 = r2_score(y_test_rating, y_pred_rating)
print("Mean Square Error: ",mse)
print("R squared: ",r2)
```

```python
feature_importances = pd.DataFrame(dtr.feature_importances_,
                                   index = X_train_rating.columns,
                                   columns=['importances']).sort_values('importances', ascending=False)
feature_importances
```

```python
#task 2: Recommendation System
def recommend_restaurant(cuisine,price_range,top_n=5):
    recommendations = df[
        (df['Cuisines'].str.contains(cuisine, case=False))&(df['Price range'] == price_range)
    ]
    recommendations = recommendations.sort_values(by=['Aggregate rating', 'Votes'], ascending=False)

    if recommendations.empty:
        return "No restaurants found matching your criteria."
    else:
        return recommendations.head(top_n)[['Restaurant Name', 'Cuisines', 'Price range', 'Aggregate rating', 'Address']]
#example
recommend_restaurant('Italian',3)
```

```python
#Task 3 Cuisine Classification
#using primary cuisine
df['Primary Cuisine'] = df['Cuisines'].apply(lambda x: x.split(',')[0] if isinstance(x, str) else x)

features_cuisine = ['Price range', 'Votes', 'Aggregate rating']
target_cuisine = 'Primary Cuisine'


```

```python
#top 10 frequent cuisines
top_cuisines = df['Primary Cuisine'].value_counts().nlargest(10).index
df_cuisine_filtered = df[df['Primary Cuisine'].isin(top_cuisines)]
```

```python
#encoding   
le_cuisine = LabelEncoder()
df_cuisine_filtered.loc[:, 'Primary Cuisine Encoded'] = le_cuisine.fit_transform(df_cuisine_filtered['Primary Cuisine'])
```

```python
X_cuisine = df_cuisine_filtered[features_cuisine]
y_cuisine = df_cuisine_filtered['Primary Cuisine Encoded']
X_train_cuisine, X_test_cuisine, y_train_cuisine, y_test_cuisine = train_test_split(X_cuisine, y_cuisine, test_size=0.2, random_state=42, stratify=y_cuisine)
```

```python
#model training using Random Forest Classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train_cuisine, y_train_cuisine)

```

```python
#model evaluation
y_pred_cuisine = classifier.predict(X_test_cuisine)
accuracy = accuracy_score(y_test_cuisine, y_pred_cuisine)
```

```python
# Use average='weighted' for multi-class classification
precision = precision_score(y_test_cuisine, y_pred_cuisine, average='weighted', zero_division=0)
recall = recall_score(y_test_cuisine, y_pred_cuisine, average='weighted', zero_division=0)
print(f"\nAccuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
print("\nPerformance by Cuisine:")
cuisine_mapping = dict(zip(le_cuisine.transform(le_cuisine.classes_), le_cuisine.classes_))
for label, cuisine_name in cuisine_mapping.items():
    cuisine_mask_true = y_test_cuisine == label
    cuisine_mask_pred = y_pred_cuisine == label
    precision_cuisine = precision_score(cuisine_mask_true, cuisine_mask_pred, zero_division=0)
    recall_cuisine = recall_score(cuisine_mask_true, cuisine_mask_pred, zero_division=0)

    print(f"- {cuisine_name}: Precision={precision_cuisine:.2f}, Recall={recall_cuisine:.2f}")
```
